```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
from matplotlib.ticker import FixedLocator

X = [750, 800, 850]

X_750C = [0/12, 1/12, 2/12, 3/12, 4/12, 5/12, 6/12, 7/12, 8/12, 9/12, 10/12]
Y_750C = [2.68173E-14, 3.73284E-14, 2.27267E-14,2.76316E-14,2.34579E-14, 2.909E-14, 2.38793E-14, 2.1207E-14, 2.1207E-14,1.94247E-14,2.23591E-14]

Y_282 = [2.2727E-14, None, 3.141E-14]  # No value for 800
Y_552 = [2.90917E-14, 2.40046E-14, 2.145E-14]
Y_822 = [2.1207E-14, 2.227E-14, 2.045E-14]

colors_750C = ['darkblue', 'darkblue','darkblue','darkblue','darkblue','darkblue','darkblue','white','darkblue','darkblue', 'darkblue']
colors_750 = ['darkblue', 'darkblue','darkblue']
colors_800 = ['red', 'red', 'red']
colors_850 = ['darkgreen', 'darkgreen', 'darkgreen']

fig, axs = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

slope, intercept, r_value, _, _ = linregress(X_750C, Y_750C)

r_squared = r_value ** 2
x_line = np.linspace(min(X_750C), max(X_750C), 100)

y_line = slope * x_line + intercept

axs[0].scatter(X_750C, Y_750C, c=colors_750C, marker='o', s=50, label='750C')
axs[0].plot(x_line, y_line, color='darkblue', linestyle='--', linewidth=2)
axs[0].set_xlabel('x in Na$_0$$_.$$_8$$_3$Fe$_x$Mn$_($$_1$$_-$$_x$$_)$O$_2$', fontsize=20)
axs[0].set_ylabel('$D_{Na^{+}}$ ($10^{-14}$)', fontsize=20)
axs[0].set_xticks(ticks=X_750C)
axs[0].set_xticklabels(['0/12', '1/12', '2/12', '3/12', '4/12', '5/12', '6/12', '7/12', '8/12', '9/12', '10/12'])
axs[0].tick_params(axis='x', which='major', labelsize=15)
axs[0].tick_params(axis='y', which='major', labelsize=16)
equation_text = f'y = {slope:.2e}x + {intercept:.2e}\n$R^{2}$ = {r_squared:.2f}'
axs[0].text(0.48, 0.98, equation_text, transform=axs[0].transAxes, fontsize=16, verticalalignment='top')

axs[0].yaxis.set_major_locator(FixedLocator(axs[0].get_yticks()))
axs[0].set_yticklabels([f'{y:.2e}'.replace('e-14', '') for y in axs[0].get_yticks()])

axs[1].scatter(X, Y_282, marker='o', s=60, label='282', color='darkseagreen')
axs[1].scatter(X, Y_552, marker='^', s=60, label='552', color='cornflowerblue')
axs[1].scatter(X, Y_822, marker='s', s=60, label='822', color='hotpink')
axs[1].set_xlabel('Temperature (C)', fontsize=20)
axs[1].set_xticks(X)
axs[1].legend(fontsize=20)
axs[1].tick_params(axis='x', which='major', labelsize=18)

plt.tight_layout()

output_file = "/Users/harryfyjis-walker/Desktop/Part II Stuff/figures/Diffusioncoefficients.pdf"
plt.savefig(output_file, format='pdf', bbox_inches='tight') 

plt.show()
```
