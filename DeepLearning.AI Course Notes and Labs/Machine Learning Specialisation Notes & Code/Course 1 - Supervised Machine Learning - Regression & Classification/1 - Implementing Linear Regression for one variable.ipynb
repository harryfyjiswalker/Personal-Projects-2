```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Predicting house price from data on house size in feet^2 and house price
# x_train is the input variable (size in 1000 sqft)
# y_train is the target/ output (price in 1000s of dollars)

#Step 1: store data in 1D NumPy arrays
x_train = np.array([1.0, 2.0]) #this is a tuple that contains the dimensions of the NumPy array x; if x is a 2D array, x.shape will return a tuple in the form (numberofrows, numberofcolumns)
y_train = np.array([300.00, 500.0])

#Step 2: let m be the total number of training examples, ...
#...where the index of a training example is denoted by i
m = len(x_train)

#Step 3: choose values of w and b
w = 200
b = 100

#Step 4: the model for linear regression (a fn that maps from x)...
#... to y is f = wx + b. Use a for loop to loop this through...
#... each data point
def compute_model_output(x, w, b): #define a function
    m = x.shape[0] #determines how many data samples (m) you have in your input data x.
                    #[0] refers to first dimension of the array x i.e. the number of rows
    f_wb = np.zeros(m) #creates a 1D array of zeros with length m, to be later filled with the model output for each sample
    for i in range(m): #each iteration will compute the model output for one sample and store it in f_wb[i]
        f_wb[i] = w * x[i] + b #x[i] is the ith row of matrix x, which represents the ith input sample. Here we compute f_wb for each input sample i
    
    return f_wb #returns the array f_wb, which contains the model outputs for all m input samples

#Step 5: call the computer_model_output function and plot the output
tmp_f_wb = compute_model_output(x_train, w, b) #when called, calculates the output for each row; the result will be assigned to the variable tmp_f_wb

plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction') # Plot our model prediction
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values') # Plot the data points

plt.title("Housing Prices")# Set the title
plt.ylabel('Price (in 1000s of dollars)') # Set the y-axis label
plt.xlabel('Size (1000 sqft)')# Set the x-axis label
plt.legend()
plt.show()

#Step 6: use this regression to make a prediction
x_i = 1.2
cost_1200sqft = w * x_i + b
print(f"${cost_1200sqft:.0f} thousand dollars")

```
